# PREDICT NUMBER OF DIVISORS

Is it possible to predict amount of divisors of a given integer with a simple machine learning model?

```python
import matplotlib.pyplot as plt

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from tqdm.notebook import tqdm
```

Define a function to compute all divisor including $ 1 $ from a given integer

```python
def compute_divisor(x):
    ds = [i for i in range(1, int(x/2)+1) if x%i == 0]
    return ds
```

### Generate input data

```python
X = [[i] for i in range(10000, 50000)]
y = []
for v in tqdm(X):
    y.append(len(compute_divisor(v[0])))
```

```python
plt.figure(figsize=(20, 10))
plt.plot(y)
```

### Train random forrest model

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
reg = RandomForestRegressor(random_state=42, n_jobs=-1)
reg.fit(X_train, y_train)
reg.score(X_test, y_test)
```

### Model evaluation

```python
y_pred = reg.predict(X_test)

plt.figure(figsize=(20, 10))
plt.plot(y_test)
plt.plot(y_pred)
```
