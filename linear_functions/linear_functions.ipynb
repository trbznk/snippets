```python
import matplotlib.pyplot as plt
from sympy import symbols
from sympy import Lambda


def cartesian(ax):
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    return ax


def interpolate(function, start, stop):
    X = [x for x in range(start, stop)]
    Y = [function(x) for x in X]
    return X, Y
```

# Linear Functions

The formal defintion of a linear function is $ f(x)=mx+b $, where $ m $ is the magnitude and $ b $ the intercept with the $ y $ axis.

Let's define a function $ f(x)= 2x + 7 $. By setting the value of $ x $, we can compute the value of the function. For example with $ x = 1 $ like:

$$ 
\begin{align*}
f(1) &= 2 \cdot 1 + 7 \\
f(1) &= 9
\end{align*}
$$

```python
m, b = 2, 7

f = lambda x: m*x+b
```

```python
y = f(1)
y
```

We can create a table of values by compute more results of $ f $.

```python
print("x", "f(x)")
print(0, f(0))
print(1, f(1))
print(2, f(2))
print(3, f(3))
print(4, f(4))
```

When visualize these points in a coordinate system, we can assume that all points lie on a straight line

```python
X = [x for x in range(5)]
Y = [f(x) for x in X]

fig, ax = plt.subplots(figsize=(10, 5))
ax = cartesian(ax)

ax.set_title("$ f(x)=2x+7 $")
ax.set_xlim(-2, 6)
ax.set_ylim(-2, 17)
ax.plot([-20, 20], [f(-20), f(20)], "orange")
ax.plot(X, Y, "b.")
plt.show()
```

## Magnitude

The maginitude has the formale definition $ m = \frac{\Delta y}{\Delta x}  $.

```python
nrows, ncols = 2, 3
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16, 10))
fig.suptitle("Visualization of the function $ f(x) = mx+2 $ with different values for the magnitude $ m $")
M = [
    [1, 1/10, 0, -10],
    [-10, -1, -1/10]
]

for row in range(nrows):
    for col in range(ncols):
        m = M[row][col]
        axs[row][col] = cartesian(axs[row][col])
        axs[row][col].set_title(f"$ m={m} $")
        lim = 20
        axs[row][col].set_xlim(-lim, lim)
        axs[row][col].set_ylim(-lim, lim)
        f = lambda x: m*x+1
        axs[row][col].plot([-lim, lim], [f(-lim), f(lim)])
```

```python
m, x, b = sympy.symbols("m x b")
m, x, b
```

```python
f = Lambda(x, m*x+b)
f
```

```python
g = f(f(x))
g
```

```python
f = Lambda(x, 2*x+7)
f
```

```python
g = Lambda(x, f(f(x)))
g
```

```python
fx, fy = interpolate(f, -10, 10)
gx, gy = interpolate(g, -10, 10)

fig, ax = plt.subplots(figsize=(12, 8))
ax = cartesian(ax)
ax.plot(fx, fy, label="$G_f$")
ax.plot(gx, gy, label="$G_g$")
plt.legend()
plt.show()
```
