# Handling Cyclical Features

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Problem with difference

The difference between hour `0` and `23` is large, even the values are near on the clock.

```python
hours = np.arange(24)
hours
```

```python
plt.plot(hours)
```

### Solution

We use sinus and cosinus tranformation to solve the issue with the difference between values.

```python
hours_sin = np.sin((2*np.pi*hours)/24)
hours_cos = np.cos((2*np.pi*hours)/24)

plt.plot(hours_sin)
plt.plot(hours_cos)
```

```python
fig, ax = plt.subplots()
ax.scatter(hours_sin, hours_cos)

for i, hour in enumerate(hours):
    ax.annotate(hour, (hours_sin[i], hours_cos[i]))
```
