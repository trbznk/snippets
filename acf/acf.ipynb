```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("melbourne_temperatures.csv", index_col=0)
df.plot(figsize=(20, 10))
```

```python
fig, axs = plt.subplots(2, figsize=(20, 10))
_ = sm.graphics.tsa.plot_acf(df["Temp"], lags=14, ax=axs[0])
_ = sm.graphics.tsa.plot_acf(df["Temp"].diff().dropna(), lags=14, ax=axs[1])
```
