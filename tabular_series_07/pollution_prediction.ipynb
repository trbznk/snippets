```python
import pandas as pd
import torch

from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('train.csv', parse_dates=['date_time'])
train
```

```python
train.describe()
```

```python
test = pd.read_csv('test.csv', parse_dates=['date_time'])
test
```

```python
train['weekday'] = train['date_time'].dt.weekday
train['hour'] = train['date_time'].dt.hour
test['weekday'] = test['date_time'].dt.weekday
test['hour'] = test['date_time'].dt.hour
train, test
```

```python
from sklearn.preprocessing import StandardScaler

train = train.drop(['date_time'], axis=1) # Parameter ais ist Zeile gleich 0, Spalten gleich 1
test = test.drop(['date_time'], axis=1)

x_cols = [col for col in train.columns if 'target' not in col]
y_cols = [col for col in train.columns if 'target' in col]

scaler = StandardScaler()
train[x_cols] = scaler.fit_transform(train[x_cols])
test[x_cols] = scaler.transform(test[x_cols])
train, test
```

```python
train, dev = train_test_split(train, test_size=0.30, random_state=42) # dev für validation set
train.shape, dev.shape
```

```python
from torch.utils.data import Dataset, DataLoader


class Data(Dataset):
    def __init__(self, df, is_test=False): # Constructor
        self.df = df
        self.is_test = is_test
        self.x_cols = [col for col in self.df.columns if 'target' not in col]
        self.y_cols = [col for col in self.df.columns if 'target' in col]

    def __len__(self): # magic method für Listen (anzahl der Elemente von Dataset)
        return len(self.df)
    
    def __getitem__(self, idx): # muss indizierbar sein
        return torch.tensor(self.df[self.x_cols].iloc[idx]).float(), torch.tensor(self.df[self.y_cols].iloc[idx]).float()
    
train_data = Data(train)
dev_data = Data(dev)
test_data = Data(test, is_test=True) # wenn wir test übergeben ist es True!
train_data, test_data, dev_data
```

```python
train_data[0] # Ausgabe sind genau zwei eindim Arrays
```

```python
data_loader = {
    'train': DataLoader(train_data, batch_size=64),
    'dev': DataLoader(dev_data, batch_size=64),
    'test': DataLoader(test_data, batch_size=64)
}
```

```python
data_loader['train']
```

```python
import torch.nn as nn

class Net(nn.Module): # Vererbung, daher Quelle in die Klammern
    
    def __init__(self):
        super().__init__()# gibt die Superklasse nn.module, außerdem rufen wir den Constuctor dieser Klasse auf!
        self.l1 = nn.Linear(10, 16)
        self.l2 = nn.Linear(16, 16)
        self.l3 = nn.Linear(16, 3)
        self.g = nn.ReLU()
    
    def forward(self, x):
        x = self.l1(x)# wir überschreiben x mit dem Ergebnis aus Layer1
        x = self.g(x)
        x = self.l2(x)
        x = self.g(x)
        x = self.l3(x)
        return x
model = Net() # wir erzeugen ein Objekt der Klasse ! und packen es in eine variable (model)
model
```

## Fitting Cycle

```python
import torch.optim as optim


class RMSLELoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.mse = nn.MSELoss()
        
    def forward(self, pred, actual):
        return torch.sqrt(self.mse(torch.log(pred+1), torch.log(actual+1)))


criterium = RMSLELoss()
optimizer = optim.Adam(model.parameters())


for epoch in range(100):
    for phase in ['train', 'dev']:
        if phase == 'train':
            model.train()
        else:
            model.eval()
        running_loss = 0.0
        for inputs, outputs in data_loader[phase]:
            optimizer.zero_grad()
            preds = model(inputs) # forwarding
            loss = criterium(preds, outputs) # loss calculation
            running_loss += loss.item()
            if phase == 'train':
                loss.backward() # backpropagation
                optimizer.step() # update weights
        print(f'{epoch+1} {phase} {(running_loss/len(data_loader[phase])):.2f}')
                
      
```
