# TEMPLATES for img2pimg

Using picasso paintings to generate different images from templates

```python
import pickle

from src import Img2Pimg

with open("templates/german.pickle", "rb") as f:
    GERMANY_TEMPLATE = pickle.load(f)

with open("templates/python.pickle", "rb") as f:
    PYTHON_TEMPLATE = pickle.load(f)
    
with open("templates/heart.pickle", "rb") as f:
    HEART_TEMPLATE = pickle.load(f)
```

```python
i2p = Img2Pimg("picasso", GERMANY_TEMPLATE, square_size=100, step_size=100, th=50)
i2p.run()
i2p.img_.save("examples/germany.png")
i2p.img_
```

```python
i2p = Img2Pimg("picasso", PYTHON_TEMPLATE, th=60)
i2p.run()
i2p.img_.save("examples/python.png")
i2p.img_
```

```python
i2p = Img2Pimg("picasso", HEART_TEMPLATE, th=60)
i2p.run()
i2p.img_.save("examples/heart.png")
i2p.img_
```
